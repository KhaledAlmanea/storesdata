```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Stores.csv')
df.head(10)
```

```python
df.info()
```

```python
import seaborn as sns
```

## Is the store area correlated with the number of items available ?

```python
sns.jointplot(x='Store_Area', y='Items_Available', data=df, kind="reg", truncate=False, color="b", height=7);
```

Both variables data are normally distributed and have high positive correlation

## Is the store area linked to the number of daily customers ?

```python
sns.jointplot(x='Store_Area', y='Daily_Customer_Count', data=df, kind="reg", truncate=False, color='b', height=7);
```

both variables data are normalliy distrebuted and has no correlation to each others.

## Do you spend more in a big store?

```python
sns.jointplot(x='Store_Area', y='Store_Sales', data=df, kind="reg", truncate=False, color='b', height=7);
```

both variables data are normalliy distrebuted and has no correlation to each others

## The more available items the more daily customers ?

```python
sns.jointplot(x='Items_Available', y='Daily_Customer_Count', data=df, kind='reg', truncate=False, color='b', height=7);
```

both variables data are normalliy distrebuted and has no correlation to each others.

## More items available, bigger invoices ?

```python
sns.jointplot(x='Items_Available', y='Store_Sales', data=df, kind='reg', truncate=False, color='b', height=7);
```

No corroltation.

## The number of daily customers has an impact on the store sales ?

```python
sns.jointplot(x='Daily_Customer_Count', y='Store_Sales', data=df, kind='reg', truncate=False, color='b', height=7);
```

No Corrolation.

## Conclusion:
The bigger the store the more items you have, but it does not mean more sales and profit margin eventully.
